Bibliotecas:

```python
import pandas as pd
import pandera as pa
```

## Extração:

```python
df = pd.read_json("DataSetTheNews.json")
df
```

Tipos de Dados:

```python
df.dtypes
```

Converter Datas:

```python
df['Entry Date'] = pd.to_datetime(df['Entry Date'], format = "%d/%m/%Y")
df['Last Opened'] = pd.to_datetime(df['Last Opened'], format = "%d/%m/%Y")
df.head()
```

Criando Colunas:

```python
df['fullName'] = df['first_name'] + ' ' + df['last_name']
df['fullName']
```

Validação:

```python
df.dtypes
```

```python
schema = pa.DataFrameSchema(
    columns= {
'first_name':pa.Column(pa.String, nullable=True),
'last_name': pa.Column(pa.String, nullable=True),
'email': pa.Column(pa.String, nullable=True),
'Entry Date': pa.Column(pa.DateTime, nullable=True),
'Last Opened': pa.Column(pa.DateTime, nullable=True),
'Open Rate': pa.Column(pa.Float64, nullable=True),
'Newsletter': pa.Column(pa.String, nullable=True),
'Entry Channel': pa.Column(pa.String, nullable=True),
'fullName': pa.Column(pa.String, nullable=True)
}
    )
```

```python
schema.validate(df)
```

## Limpeza:

Verificar Dados não Informados - Sumarizado:

```python
df.isnull().sum()
```

Excluir registros com Dados não Informados:

```python
df.dropna(subset=['last_name'], inplace=True)
df.dropna(subset=['first_name'], inplace=True)
```
